```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
copy_sc=pd.read_csv('C:/Users/SHIVANSHU SRIVASTAVA/OneDrive/Desktop/ml by upflair/department_salary.csv')
```

```python
copy_sc
```

```python
# # Plot a bar chart to visualize the number of employees in each department using Matplotlib.

copy_sc['Department'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
```

```python
# Filter out employees with a salary greater than 60,000 using Pandas

high_salary_employees = copy_sc[copy_sc['Salary'] > 60000]
print(high_salary_employees)
```

```python
#  Create a line graph for the salary trend by department.

copy_sc.groupby('Department')['Salary'].mean().plot(kind='line', marker='o', color='orange')
plt.title('Average Salary Trend by Department')
```
