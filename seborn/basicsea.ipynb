```python
sns.relplot(x='total_bill', y='tip', data=data, hue='time')# we are doing scatter plot of total bill and tip with hue time
# hue is used to show the different color in the plot
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=data)# we are doing scatter plot of total bill and tip with hue time and style smoker
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=data,kind='scatter',col='smoker')# we are doing scatter plot of total bill and tip with hue time and style smoker and col smoker
# we are doing scatter plot of total bill and tip with hue time and style smoker and col smoker
```

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',style='smoker',col='smoker',row='time')# we are doing scatter plot of total bill and tip with hue time and style smoker and col smoker
```

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='time')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill',color='r')
plt.ylabel('Tip',color='green',fontsize=20)
```
